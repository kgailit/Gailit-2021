```python
import re
from estnltk.names import LAYER_CONLL
from estnltk import Text
from pprint import pprint
from pathlib import Path
import os
```

```python
esimesed = []

with open("etnc19_web_2019.prevert", "r", encoding = "UTF8") as korpus:
    arv = 0
    while True:
        rida = korpus.readline()
        if rida[0] == "<":
            continue
        arv += 1
        esimesed.append(rida)
        if arv == 10:
            break
```

```python
tekst = Text(" ".join(esimesed))
tekst.tag_syntax()

pprint( list(zip(tekst[LAYER_CONLL], tekst.word_texts)) )
```

```python
import stanza

nlp = stanza.Pipeline(lang='et', processors='tokenize,mwt,pos,lemma,depparse')
doc = nlp(" ".join(esimesed))
for sent in doc.sentences:
    for word in sent.words:
        print(word)
```
